### Amazon

```python
df_amazon1 = pd.read_json('../../datasets/amazon_tv.json')
df_amazon2 = pd.read_json('../../datasets/amazon_detailed.json')
df_amazon3 = pd.read_json('../../datasets/amazon1.json')

df_amazon1.shape, df_amazon2.shape, df_amazon3.shape
```

```python
set(list(df_amazon3)) - set(list(df_amazon2))
```

```python
set(list(df_amazon3)) - set(list(df_amazon1))
```

```python
set(list(df_amazon2)) - set(list(df_amazon1))
```

```python
df_amazon3 = df_amazon3[df_amazon3.error_code.isna()].reset_index(drop=True)
```

```python
amazon_columns = list(df_amazon1.columns)
```

```python
df_amazon1[amazon_columns].shape, df_amazon2[amazon_columns].shape, df_amazon3[amazon_columns].shape
```

```python
df_amazon = pd.concat([df_amazon1[amazon_columns], df_amazon2[amazon_columns], df_amazon3[amazon_columns]], axis=0).reset_index(drop=True)
```

```python
df_amazon.title.drop_duplicates().shape
```

```python
df_amazon.department.value_counts()
```

```python
df_amazon.to_json('../../datasets/amazon_final.json', orient='records', lines=True)
```

### Target

```python
df_target1=  pd.read_json('../../datasets/target_tv.json')
df_target2 =  pd.read_json('../../datasets/target1.json')
df_target1.shape, df_target2.shape
```

```python
target_common_columns = list(set(df_target1) & set(df_target2))
```

```python
df_target1 = df_target1[df_target1.error_code.isna()].reset_index(drop = True)
df_target2 = df_target2[df_target2.error_code.isna()].reset_index(drop = True)
```

```python
df_target = pd.concat([df_target1[target_common_columns], df_target2[target_common_columns]], axis=0).reset_index(drop=True)
```

```python
df_target.head()
```

```python
df_target.to_json('../../datasets/target_final.json', orient='records', lines=True)
```
